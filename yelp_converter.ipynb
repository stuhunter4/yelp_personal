```python
import pandas as pd
import json
```

```python
df = pd.read_csv("FremontCA_Restaurants_17_02_2021.csv")
df
```

```python
# drop columns we won't be displaying on the web page
clean_df = df.drop(["id", "alias", "is_closed", "country", "state"], axis = 1)
```

```python
# fix columns with [] using string replace
clean_df["transactions"] = clean_df["transactions"].str.replace("[", "")
clean_df["transactions"] = clean_df["transactions"].str.replace("]", "")

clean_df["display_address"] = clean_df["display_address"].str.replace("[", "")
clean_df["display_address"] = clean_df["display_address"].str.replace("]", "")
```

```python
# display the new dataframe
clean_df
```

```python
# use .to_dict and 'records' orient to make our desired list
# can be used as 'data.js', or .insert_many with PyMongo
list_d = clean_df.to_dict(orient='records')
```

```python
# write to a text file for our data.js later
with open("FremontCA_Restaurants_17_02_2021.txt", "w") as file:
    file.write(json.dumps(list_d))
```
